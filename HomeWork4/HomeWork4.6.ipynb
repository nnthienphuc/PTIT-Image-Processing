```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


def read_binary_image(filename, shape=(256, 256)):
    with open(filename, 'rb') as f:
        img = np.fromfile(f, np.uint8)
    return img.reshape(shape)

def display_image(data, title):
    plt.imshow(data, cmap='gray')
    plt.title(title)
    plt.axis('off')
    plt.colorbar()
    plt.show()


def process_image_dft(image, filename):
    """Compute and display DFT components of the image."""
    # Compute the DFT and center it
    dft = np.fft.fftshift(np.fft.fft2(image))

    # Real and Imaginary parts
    real_part = np.real(dft)
    imag_part = np.imag(dft)

    # Log-Magnitude Spectrum
    magnitude_spectrum = np.log(np.abs(dft) + 1)  # add 1 to avoid log(0)

    # Phase
    phase = np.angle(dft)

    # Display the results
    display_image(image, "Original Image")
    display_image(real_part, "Real Part of DFT")
    display_image(imag_part, "Imaginary Part of DFT")
    display_image(magnitude_spectrum, "Log-Magnitude Spectrum of DFT")
    display_image(phase, "Phase of DFT")

    cv.imwrite(f"D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/(4_6)Original Image {filename}.png", image)
    cv.imwrite(f"D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/(4_6)Real Part of DFT {filename}.png", real_part)
    cv.imwrite(f"D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/(4_6)Imaginary Part of DFT {filename}.png", imag_part)
    cv.imwrite(f"D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/(4_6)Log-Magnitude Spectrum of DFT {filename}.png", magnitude_spectrum)
    cv.imwrite(f"D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/(4_6)Phase of DFT {filename}.png", phase)


# Load images and process each
image_files = ['camerabin.sec', 'salesmanbin.sec', 'headbin.sec', 'eyeRbin.sec']
for filename in image_files:
    image = read_binary_image('D:/Full-Document/Information-Technology/Image-Processing/Exercises/HomeWork4/' + filename)
    process_image_dft(image, filename)
```
