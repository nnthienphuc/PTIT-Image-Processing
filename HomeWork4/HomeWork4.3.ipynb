```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


def read_binary_image(image_path):
    with open(image_path, 'rb') as f:
        return np.fromfile(f, np.uint8)


cols, rows = np.meshgrid(np.arange(8), np.arange(8))
# Problem 1: Define I3
u0 = 2
v0 = 2
I3 = np.cos(1j * 2 * np.pi / 8 * (u0 * cols + v0 * rows))
# Real and Imaginary parts as grayscale images
plt.subplot(1,2,1)
plt.imshow(I3.real, cmap='gray')
plt.title("Real Part of I3")
cv.imwrite("Homework4/(4_3)Real Part of I3.png", I3.real)


plt.subplot(1,2,2)
plt.imshow(I3.imag, cmap='gray')
plt.title("Imaginary Part of I3")
cv.imwrite("Homework4/(4_3)Imaginary Part of I3.png", I3.imag)

plt.show()

# Compute centered DFT of I3
I3_dft = np.fft.fftshift(np.fft.fft2(I3))

# Print real and imaginary parts of the DFT
print("Real Part of DFT(I3):\n", np.round(I3_dft.real, 4))
print("Imaginary Part of DFT(I3):\n", np.round(I3_dft.imag, 4))
```
