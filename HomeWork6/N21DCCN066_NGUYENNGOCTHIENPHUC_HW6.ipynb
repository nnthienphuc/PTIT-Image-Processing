```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import cv2
import os

output_dir = "result"
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

def read_sec(file_path, shape=(256, 256)):
    with open(file_path, 'rb') as file:
        data = np.fromfile(file, dtype=np.uint8)
    return data.reshape(shape)

def median_filter(image):
    return cv2.medianBlur(image, 3)
def morphological_operation(image, operation):
    kernel = np.ones((3, 3), np.uint8)
    if operation == 'open':
        return cv2.morphologyEx(image, cv2.MORPH_OPEN, kernel)
    elif operation == 'close':
        return cv2.morphologyEx(image, cv2.MORPH_CLOSE, kernel)
    else:
        raise ValueError("Invalid morphological operation")

camera9bin = read_sec('camera9bin.sec')
camera99bin = read_sec('camera99bin.sec')

camera9bin_median = median_filter(camera9bin)[1:-1, 1:-1]
camera9bin_open = morphological_operation(camera9bin, 'open')[1:-1, 1:-1]
camera9bin_close = morphological_operation(camera9bin, 'close')[1:-1, 1:-1]

camera99bin_median = median_filter(camera99bin)[1:-1, 1:-1]
camera99bin_open = morphological_operation(camera99bin, 'open')[1:-1, 1:-1]
camera99bin_close = morphological_operation(camera99bin, 'close')[1:-1, 1:-1]

plt.figure(figsize=(15, 10))

plt.subplot(2, 4, 1)
plt.imshow(camera9bin, cmap='gray')
plt.title('Original Image (camera9bin)')

plt.subplot(2, 4, 2)
plt.imshow(camera9bin_median, cmap='gray')
plt.title('Median Filter')

plt.subplot(2, 4, 3)
plt.imshow(camera9bin_open, cmap='gray')
plt.title('Morphological Opening')

plt.subplot(2, 4, 4)
plt.imshow(camera9bin_close, cmap='gray')
plt.title('Morphological Closing')

plt.subplot(2, 4, 5)
plt.imshow(camera99bin, cmap='gray')
plt.title('Original Image (camera99bin)')

plt.subplot(2, 4, 6)
plt.imshow(camera99bin_median, cmap='gray')
plt.title('Median Filter')

plt.subplot(2, 4, 7)
plt.imshow(camera99bin_open, cmap='gray')
plt.title('Morphological Opening')

plt.subplot(2, 4, 8)
plt.imshow(camera99bin_close, cmap='gray')
plt.title('Morphological Closing')

plt.savefig(os.path.join(output_dir, "HomeWork6.png"))
plt.show()

print(f"\nẢnh đã được lưu vào thư mục: {output_dir}")
```
