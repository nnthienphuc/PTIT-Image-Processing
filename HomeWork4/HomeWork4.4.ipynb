```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


def read_binary_image(image_path):
    with open(image_path, 'rb') as f:
        return np.fromfile(f, np.uint8)


cols, rows = np.meshgrid(np.arange(8), np.arange(8))
# Problem 1: Define I4
u0 = 2
v0 = 2
I4 = np.sin(1j * 2 * np.pi / 8 * (u0 * cols + v0 * rows))
# Real and Imaginary parts as grayscale images
plt.subplot(1,2,1)
plt.imshow(I4.real, cmap='gray')
plt.title("Real Part of I4")
cv.imwrite("Homework4/(4_4)Real Part of I4.png", I4.real)

plt.subplot(1,2,2)
plt.imshow(I4.imag, cmap='gray')
plt.title("Imaginary Part of I4")
cv.imwrite("Homework4/(4_4)Imaginary Part of I4.png", I4.imag)

plt.show()

# Compute centered DFT of I4
I4_dft = np.fft.fftshift(np.fft.fft2(I4))

# Print real and imaginary parts of the DFT
print("Real Part of DFT(I4):\n", np.round(I4_dft.real, 4))
print("Imaginary Part of DFT(I4):\n", np.round(I4_dft.imag, 4))
```
